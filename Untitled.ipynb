```python
# -*- coding: utf-8 -*-
from transformers import AutoTokenizer, AutoModel
import os

# 这里是模型下载的位置
model_dir = 'C:\\Users\\25028\Desktop\\2024s1\\COMP8851MajorProject\\MainProject\\comp8851project\\llms\\chatglm2-6b'

tokenizer = AutoTokenizer.from_pretrained(model_dir, trust_remote_code=True)
model = AutoModel.from_pretrained(model_dir, trust_remote_code=True)
history = []
while True:
    query = input("\n用户：")
    if query == "stop":
        break
    if query == "clear":
        history = []
        os.system('clear')
        continue
    response, history = model.chat(tokenizer, query, history=history)
    print(f"小元：{response}")
```
